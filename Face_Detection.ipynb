```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
```

```python
face = DeepFace.detectFace(
    "gourav_image.png", target_size=(224,224),detector_backend = "opencv"
)
```

```python
plt.imshow(face[0])  # Plot the face
plt.axis('off')  # Turn off axis
plt.show()  # Show the face
```
