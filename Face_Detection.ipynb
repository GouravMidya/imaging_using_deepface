```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
```

```python
backends = ["opencv", "ssd", "mtcnn", "retinaface"]
```

```python
face = DeepFace.detectFace(
    "gourav_image.png", target_size=(224,224),detector_backend = "opencv"
)
```

```python
plt.imshow(face[0])  # Plot the face
plt.axis('off')  # Turn off axis
plt.show()  # Show the face
```

Trying out multiple backends

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs = axs.flatten()

for i, b in enumerate(backends):
    try:
        axs[i].set_title(b)
        face = DeepFace.detectFace("gourav_image.png", target_size=(224,224), detector_backend=b)
        axs[i].imshow(face[0])  
        axs[i].axis("off")
    except Exception as e:
        print(f"Error with backend {b}: {str(e)}")
        pass

plt.show()
```

```python
face1 = DeepFace.detectFace(
    "gourav_image.png", target_size=(224,224),detector_backend = "opencv"
)
face2 = DeepFace.detectFace(
    "gourav_crop.jpg", target_size=(224,224),detector_backend = "opencv"
)

# Create a figure with one row and two columns for the subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first face in the first subplot
axs[0].imshow(face1[0])
axs[0].set_title('Face 1')
axs[0].axis('off')

# Plot the second face in the second subplot
axs[1].imshow(face2[0])
axs[1].set_title('Face 2')
axs[1].axis('off')

plt.show()
```
